# - Different Algorithms Applied on Data Set

```python
X = df.iloc[:, 1:].values
Y = df.iloc[:, 0:1].values
print X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.33, random_state = 0)
print X_train.shape, Y_train.shape
print X_test.shape, Y_test.shape
```

                                            (LogisticRegression Algorithm)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
clf = LogisticRegression()
clf.fit(X_train, Y_train.ravel())
Y_pred = clf.predict(X_test)
accuracy_lr = round( accuracy_score(Y_pred, Y_test) * 100, 2)
print (str(accuracy_lr) + ' percent')
```

                                    (LogisticRegression Algo. With Regularisation Parameter)

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(C=0.08)#REGULARISATION PARAMETER
clf.fit(X_train, Y_train.ravel())
Y_pred_logr = clf.predict(X_test)
accuracy_lru = round( accuracy_score(Y_pred_logr, Y_test) * 100, 2)
print (str(accuracy_lru) + ' percent')
```

                                                    ( KNN Algorithm)

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 3)
clf.fit(X_train, Y_train.ravel())
Y_pred = clf.predict(X_test)
accuracy_knn = round(accuracy_score(Y_pred, Y_test) * 100, 2)
print (str(accuracy_knn) + ' percent')
```

                                             (Gaussian Naive Bayes Algorithm)

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train,Y_train.ravel())
Y_pred=clf.predict(X_test)
accuracy_gnb=round(accuracy_score(Y_pred,Y_test)*100,2)
print (str(accuracy_gnb)+ 'percent')
```

                                                 (Decision Tree Algorithm)

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,Y_train.ravel())
Y_pred=clf.predict(X_test)
accuracy_dtc=round(accuracy_score(Y_pred,Y_test)* 100,2)
print (str(accuracy_dtc) + 'percent')
```

                                                 (Random Forest Algorithm)

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train,Y_train.ravel())
Y_pred=clf.predict(X_test)
accuracy_rfc=round(accuracy_score(Y_pred,Y_test)*100,2)
print (str(accuracy_rfc)+'percent')
```

#  - Accuracy Score Table Of All Algorithms

```python
models = pd.DataFrame({
    'Algorithm_names': ['Logistic_Regression', 'KNN', 'Gaussion_NB', 'Decision_Tree', 'Random_Forest'],
    
    'Accuracy_score': [accuracy_lru,accuracy_knn,accuracy_gnb,accuracy_dtc,accuracy_rfc]
    })

models.sort_values(by='Accuracy_score', ascending=False)
```
